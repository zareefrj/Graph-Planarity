```python
import random
import networkx as nx
import matplotlib.pyplot as plt

# set the number of graphs you want to generate
num_graphs = 10

# set the number of vertices in each graph
num_vertices = 20

# loop over the number of graphs to generate
for i in range(num_graphs):
    # create a new empty graph
    G = nx.Graph()

    # add vertices to the graph
    G.add_nodes_from(range(num_vertices))

    # probability to add an edge
    if i>(num_graphs/2):
      add_edge_prob= 0.5 # 1st half of the graphs will have 50% probability of adding an edge
    else:
      add_edge_prob= 0.1 #last half will have 10% probability of adding an edge

    # loop over all pairs of vertices and add an edge with a specific probability
    for u in range(num_vertices):
        for v in range(u+1, num_vertices):
            if random.random() < add_edge_prob:
                G.add_edge(u, v)

    # print the graph number
    print(f"Graph {i+1}:")

    # check planarity
    if nx.is_planar(G):
        print("Planar")
        pos = nx.planar_layout(G)  # Use planar_layout for planar graphs
    else:
        print("Non-planar")
        pos = nx.spring_layout(G)  # Use spring_layout for non-planar graphs

    # draw the graph
    nx.draw(G, pos=pos, with_labels=True)

    # show the graph
    plt.show()
```
